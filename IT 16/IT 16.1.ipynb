```python
import pandas as pd
import numpy as np
import matplotlib
# загружаем данные
import sns as sns

data = pd.read_csv('https://raw.githubusercontent.com/FBosler/AdvancedPlotting/master/combined_set.csv')
data['Mean Log GDP per capita'] = data.groupby('Year')['Log GDP per capita'].transform(
 pd.qcut,
 q=5,
 labels=(['Lowest','Low','Medium','High','Highest'])
 )
np.exp(data[data['Year']==2018]['Log GDP per capita']).plot(
 kind='hist'
)
```

```python
data[data['Year'] == 2018].set_index('Country name')['Life Ladder'].nlargest(15).plot(kind = 'bar', figsize = (12,8))
```

```python
np.exp(data[
        data['Year'] == 2018
    ].groupby('Continent')['Log GDP per capita'].mean()).sort_values().plot(
        kind='barh',
        figsize=(12,8)
    )
```

```python
data['Life Ladder'].plot(
    kind='box',
    figsize=(12,8)
)
```

```python
data[['Healthy life expectancy at birth','Gapminder Life Expectancy']].plot(
    kind='scatter',
x='Healthy life expectancy at birth',
    y='Gapminder Life Expectancy',
    figsize=(12,8)
)
```

```python
data[data['Year'] == 2018].groupby(
    ['Continent']
)['Gapminder Population'].sum().plot(
    kind='pie',
    figsize=(12,8),
    cmap="Blues_r", # по умолчанию оранжевый
)
```

```python
data.groupby(
    ['Year','Continent']
)['Gapminder Population'].sum().unstack().plot( kind='area',
    figsize=(12,8),
# меняем цвет, если надо cmap="Blues"
)
```

```python
data[
    data['Country name'] == 'Germany'
].set_index('Year')['Life Ladder'].plot(
    kind='line',
    figsize=(12,8)
)
```

```python
import seaborn as sns
sns.jointplot(
        x='Log GDP per capita',
        y='Life Ladder',
        data=data,
        kind='scatter' # or 'kde' or 'hex'
)
```

```python
sns.jointplot(
        x='Log GDP per capita',
        y='Life Ladder',
        data=data,
        kind='hex' # or 'kde' or 'hex'
)
```
